```python
import requests
import json
import csv 
import pandas as pd
from tqdm.notebook import tqdm
import time

del(result)
result = pd.DataFrame()
```

```python
for number in tqdm(range(29024911,29200000)):
    down_url = f'https://api.kbland.kr/land-property/property/dtailInfo?%EB%A7%A4%EB%AC%BC%EC%9D%BC%EB%A0%A8%EB%B2%88%ED%98%B8={number}'
    r = requests.get(down_url,headers={
        "Accept-Encoding": "gzip",
        "Host": "api.kbland.kr",
        "Referer": "https://kbland.kr/",
        "Sec-Fetch-Dest": "empty",
        "Sec-Fetch-Mode": "cors",
        "Sec-Fetch-Site": "same-site",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36"
    })
    time.sleep(0.3)
    r.encoding = "utf-8-sig"
    temp=json.loads(r.text)
    if 'data' not in temp['dataBody']:
        continue
     
    if temp['dataBody']['data']['dtailInfo'][''] != None:
        if temp['dataBody']['data']['dtailInfo'][''] == "08" or temp['dataBody']['data']['dtailInfo'][''] == "09" or temp['dataBody']['data']['dtailInfo'][''] == "04":
            if temp['dataBody']['data']['dtailInfo'][''].find("") != -1:
                row = {}
                for i, col in enumerate(temp['dataBody']['data']['dtailInfo'].keys()):
                    row[col] = temp['dataBody']['data']['dtailInfo'][col]
                result = result.append(row,ignore_index = True)
                print(number)
```
